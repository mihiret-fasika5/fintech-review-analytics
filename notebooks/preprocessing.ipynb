```python
import pandas as pd

# Load raw dataset
df = pd.read_csv("ethiopian_banks_reviews.csv")

# Track initial size
initial_count = len(df)
print(df)
# 1. Remove duplicate reviews
df = df.drop_duplicates()
after_duplicates = len(df)

# 2. Handle missing values (review or rating required)
df = df.dropna(subset=["review text", "rating (1–5)"])
after_missing = len(df)

# 3. Normalize dates to YYYY-MM-DD
df["review date"] = pd.to_datetime(df["review date"], errors="coerce")
df = df.dropna(subset=["review date"])
df["review date"] = df["review date"].dt.strftime("%Y-%m-%d")
after_date = len(df)

# 4. Keep required columns only
df = df[["review text", "rating (1–5)", "review date", "bank / app name", "source"]]

# 5. Save cleaned dataset
output_file = "cleaned_reviews.csv"
df.to_csv(output_file, index=False)

# 6. Document cleaning stats
print("Initial rows:", initial_count)
print("After removing duplicates:", after_duplicates)
print("After dropping missing values:", after_missing)
print("After date normalization:", after_date)
print("Final rows:", len(df))
```
